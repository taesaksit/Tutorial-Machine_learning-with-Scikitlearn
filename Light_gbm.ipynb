```python
import lightgbm as lgb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# light gbm 3.1.1
```

```python
df = pd.read_csv('/Users/tae/Desktop/Project/Code/heart.csv')
df.head()
```

```python
# Split Data
from sklearn.model_selection import train_test_split
X = df.drop(columns=['target'])
y = df['target']
X_train , X_test ,y_train ,y_test =  train_test_split(X,y,test_size=0.3)
```

```python
# Create Model
model = lgb.LGBMClassifier(n_estimators=3000,max_depth=-10,learning_rate=0.5)

model.fit(X_train , y_train , 
         early_stopping_rounds=10,
         eval_metric=['auc','binary_logloss'],
         eval_set=[(X_test,y_test)],
         verbose=10)


fig, axes = plt.subplots(1, 3, figsize=(20, 5))  # Create a figure with two subplots

lgb.plot_metric(model, metric='auc', ax=axes[0] ,)
axes[0].set_title('AUC')

lgb.plot_metric(model , metric='binary_logloss' , ax=axes[1])
axes[1].set_title('Binary Log Loss')

lgb.plot_importance(model , ax=axes[2])
axes[2].set_title('Feature Importance')
```

```python
from sklearn.metrics import accuracy_score , classification_report
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
acc_train  = accuracy_score(y_train , y_pred_train)
acc_test = accuracy_score(y_test , y_pred_test)
print(f'Train: {acc_train}')
print(f'Test: {acc_test}')

print(classification_report(y_test , y_pred_test))
```
