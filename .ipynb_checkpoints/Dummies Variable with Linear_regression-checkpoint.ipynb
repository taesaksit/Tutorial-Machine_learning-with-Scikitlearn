```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/msleep.csv')
df.head()
```

```python
df.info()
```

```python
# missing value
df[df.vore.isnull()]

drop = df.dropna(subset=['vore'] , axis='index')
```

```python
drop.head(10)
```

```python
#Dummy value

dummies = pd.get_dummies(df['vore'])*1
```

```python
dummies
```

```python
#merge data 

dm = pd.concat([df,dummies] , axis=1)
dm.head()
```

```python
# Sklearn linear

from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

dt = dm[['vore','bodywt','carni','herbi','omni','insecti','sleep_total']].dropna(axis='index')
dt
```

```python
X = dt[['bodywt','herbi','omni','insecti']]
y = dt['sleep_total']
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Stat
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model_a = smf.ols(formula='sleep_total ~ + bodywt + C(vore)' , data=dt).fit()
```

```python
print(model_a.summary())
```
