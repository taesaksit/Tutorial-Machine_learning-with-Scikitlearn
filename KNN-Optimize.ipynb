```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split , RandomizedSearchCV , KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score , classification_report

path = '/Users/tae/Desktop/Project/Code/heart.csv'
df = pd.read_csv(path)
df.head()
```

```python
# Cross Validation

X = df.drop(columns=['target'])
y = df['target']
X = np.array(X)
y = np.array(y)

kf = KFold(n_splits=5)
X_train_arr = []
y_train_arr = []
X_test_arr = []
y_test_arr = []



for i,(train_index , test_index) in enumerate(kf.split(X)):
    
    X_train , X_test = X[train_index] , X[test_index]
    y_train , y_test = y[train_index] , y[test_index]

    X_train_arr.append(X_train)
    X_test_arr.append(X_test)
    y_train_arr.append(y_train)
    y_test_arr.append(y_test)
    


model = KNeighborsClassifier()
parameter = {'algorithm': ['ball_tree','kd_tree','brute'],
             'n_neighbors': np.arange(1,20),
             'weights': ['uniform','distance']}
model_optimize = RandomizedSearchCV(model , parameter)


for item in range(kf.get_n_splits()):
    
    acc_train_arr = []
    acc_test_arr = []
    knn = model_optimize.fit(X_train_arr[item] , y_train_arr[item])
    pred_train = knn.predict(X_train_arr[item])
    pred_test = knn.predict(X_test_arr[item])
    acc_train = accuracy_score(y_train_arr[item], pred_train)
    acc_test = accuracy_score(y_test_arr[item], pred_test)
    acc_train_arr.append((acc_train))
    acc_test_arr.append((acc_test))
    print(f'fold:{item+1} {acc_test} %')

print('_'*100,'\n')
print('Average train:',np.average(acc_train_arr))
print('Average test:',np.average(acc_test_arr))
    
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y , test_size=0.3)

model.fit(X_train , y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Acc_train:',accuracy_score(y_train , train_pred))
print('Acc_test:',accuracy_score(y_test , test_pred))

model_optimize.fit(X_train , y_train)
train_pred = model_optimize.predict(X_train)
test_pred = model_optimize.predict(X_test)
print('Acc_train:',accuracy_score(y_train , train_pred))
print('Acc_test:',accuracy_score(y_test , test_pred))
```
